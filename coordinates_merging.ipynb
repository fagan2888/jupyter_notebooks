# Integrating Location Hierarchies

Location relational database organized along parent-child axis. This code consolidates three levels down.

```python
import pandas as pd
```

```python
places = pd.read_csv('data/coordinates_merging/place_names.csv', names=['id', 'name', 'simple_name'], index_col='id')
places = places[['name']]
```

```python
coords = pd.read_csv('data/coordinates_merging/coords.csv', names=['id', 'lat', 'lon'], index_col='id')
```

```python
levels = pd.read_csv('data/coordinates_merging/hierarchies.csv', names=['parent', 'child']).dropna()
levels['parent'] = levels['parent'].astype(int)
```

## Inheritor Merge

```python
places0 = places.join(coords)
# pd.merge(places, coords, how='left', left_index=True, right_index=True)
```

```python
places1 = pd.merge(places0.reset_index(), levels, left_on='id', right_on='child')
places2 = places1.join(places0.add_prefix('parent_'), on='parent')
places2['lat'] = places2['lat'].fillna(places2['parent_lat'])
places2['lon'] = places2['lon'].fillna(places2['parent_lon'])
places2.groupby('id').first()

#This is one level up; possible to do a few more levels.
```
